# DIM MODEL CREATION


```python
df_src= spark.sql("SELECT DISTINCT Model_ID, mode_category FROM parquet.`abfss://silver@datalakeproject1goks.dfs.core.windows.net/car_sales`")
```

```python
df_src.display()
```

# DF SINK 

```python
if spark.catalog.tableExists('cars_catalog.gold.SCD2_dim_model'):
    df_sink= spark.sql(''' SELECT dim_model_key, Model_ID, mode_category,start_date,end_date,is_active from cars_catalog.gold.SCD2_dim_model ''')
else: 
    df_sink = spark.sql('''
        SELECT 
            1 AS dim_model_key,
            Model_ID,
            mode_category,
            CAST(NULL AS DATE) AS start_date,
            CAST(NULL AS DATE) AS end_date,
            CAST(NULL AS BOOLEAN) AS is_active
        FROM parquet.`abfss://silver@datalakeproject1goks.dfs.core.windows.net/car_sales` 
        WHERE 1=0
    ''')
```

# Filtering old and new records

```python
df_filter= df_src.join(df_sink,df_src['Model_ID']==df_sink['Model_ID'],'left').select(df_src['Model_ID'], df_src['mode_category'], df_sink['dim_model_key'],df_sink['start_date'],df_sink['end_date'],df_sink['is_active'])
```

```python
df_filter.display()
```

```python
#df_filter = df_filter.dropDuplicates(['mode_category'])
```

```python
df_filter.display()
```

**df_filter_OLD**

```python
df_filter_old= df_filter.filter(col('dim_model_key').isNotNull())
```

```python
df_filter_old.display()
```

**DF_FILTER NEW**

```python
df_filter_new= df_filter.filter(col('dim_model_key').isNull())
```

```python
df_filter_new.display()
```

# ADD SURROGATE KEY

```python
if (incremental_flag == '0'):
    max_value= 1
else:
    max_value_df = spark.sql('select max(dim_model_key) from cars_catalog.gold.SCD2_dim_model')
    max_value = (max_value_df.collect()[0][0])+1

```

# Create Surrogate Key column and add surrogate key

```python
df_filter_new= df_filter_new.withColumn('dim_model_key',max_value+ monotonically_increasing_id() )
```

```python
df_filter_new.display()
```

# Adding Dates for New DF

```python
df_filter_new = df_filter_new\
    .withColumn("start_date", current_date())\
    .withColumn("end_date", lit('9999-01-01').cast("date"))\
    .withColumn("is_active", lit(True))
```

```python
df_filter_new.display()
```

**FINAL DF**

```python
final_df= df_filter_old.union(df_filter_new)
```

```python
final_df.display()
```

# SCD Type 2

```python
if spark.catalog.tableExists("cars_catalog.gold.SCD2_dim_model"):

    delta_tbl = DeltaTable.forPath(
        spark,
        "abfss://gold@datalakeproject1goks.dfs.core.windows.net/SCD2_dim_model"
    )

    # Step 1: Expire old records that have changed
    (
        delta_tbl.alias("trg")
        .merge(
            final_df.alias("src"),
            """
            trg.Model_ID = src.Model_ID
            AND trg.is_active = true
            """
        )
        .whenMatchedUpdate(
            condition="""
                trg.mode_category <> src.mode_category
            """,
            set={
                "end_date": current_date(),
                "is_active": lit(False)
            }
        )
        .whenNotMatchedInsert(
            values={
                "dim_model_key": "src.dim_model_key",
                "mode_category": "src.mode_category",
                "Model_ID": "src.Model_ID",
                "start_date": "src.start_date",
                "end_date": "src.end_date",
                "is_active": "src.is_active"
            }
        )
        .execute()
    )

    # Step 2: Insert new versions of changed records
    # Find records that were just expired (changed today)
    changed_records = spark.sql("""
        SELECT DISTINCT Model_ID 
        FROM cars_catalog.gold.SCD2_dim_model 
        WHERE end_date = current_date() 
        AND is_active = false
    """)

    if changed_records.count() > 0:
        # Get the new versions from final_df
        new_versions = final_df.join(
            changed_records, 
            final_df.Model_ID == changed_records.Model_ID, 
            "inner"
        ).select(final_df["*"])
        
        # Get new surrogate keys for the new versions
        max_key = spark.sql('SELECT max(dim_model_key) FROM cars_catalog.gold.SCD2_dim_model').collect()[0][0]
        
        new_versions = new_versions \
            .withColumn('dim_model_key', lit(max_key + 1) + monotonically_increasing_id()) \
            .withColumn("start_date", current_date()) \
            .withColumn("end_date", lit('9999-01-01').cast("date")) \
            .withColumn("is_active", lit(True))
        
        # Append the new versions
        new_versions.write.format('delta') \
            .mode('append') \
            .option("path", 'abfss://gold@datalakeproject1goks.dfs.core.windows.net/SCD2_dim_model') \
            .saveAsTable("cars_catalog.gold.SCD2_dim_model")

else:
    final_df.write.format('delta') \
        .mode('overwrite') \
        .option("path", 'abfss://gold@datalakeproject1goks.dfs.core.windows.net/SCD2_dim_model') \
        .saveAsTable("cars_catalog.gold.SCD2_dim_model")
```
