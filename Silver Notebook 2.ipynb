# DATA READING



```python
df= spark.read.format("parquet")\
              .option('inferSchema', True)\
              . load("abfss://bronze@datalakeproject1goks.dfs.core.windows.net/rawdata") 

              
```

```python
#df.display()
```

# DATA TRANSFORMATIONS

```python
from pyspark.sql.functions import *
```

```python
df= df.withColumn("mode_category", split(col("Model_ID"),'-')[0])
#df.display()
                  
```

```python
#df.printSchema()
```

```python
df = df.withColumn("Units_Sold", col("Units_Sold").cast(StringType()))
#df.display()
```

```python
df= df.withColumn("Rev_Per_Unit", col("Revenue")/col("Units_Sold"))
```

# DATE WRITING

```python
df.write.format('parquet')\
        .mode('overwrite')\
        .option('path', 'abfss://silver@datalakeproject1goks.dfs.core.windows.net/car_sales')\
        .save()
```
